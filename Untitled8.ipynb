Write a code to generate a random sentence using probabilistic modeling
(Markov Chain). Use the sentence "The cat is on the mat" as an example

```python
import random

# Given sentence
sentence = "The cat is on the mat"
words = sentence.split()

# Build Markov Chain Model
markov_chain = {}
for i in range(len(words)-1):
    if words[i] not in markov_chain:
        markov_chain[words[i]] = []
    markov_chain[words[i]].append(words[i+1])

# Generate random sentence
current_word = random.choice(words)
generated_sentence = [current_word]

for _ in range(5):  # Generate 5 words
    next_word = random.choice(markov_chain.get(current_word, words))
    generated_sentence.append(next_word)
    current_word = next_word

print(' '.join(generated_sentence))
```

-Build a simple Autoencoder model using Keras to learn a compressed
representation of a given sentence. Use a dataset of your choice

```python
from keras.models import Model
from keras.layers import Input, Dense
import numpy as np

# Example data (simple text data)
data = ["The cat is on the mat", "The dog is in the house", "The bird is on the tree"]
data = [sentence.split() for sentence in data]  # Tokenize by words
vocab = list(set([word for sentence in data for word in sentence]))
word_to_index = {word: i for i, word in enumerate(vocab)}
index_to_word = {i: word for i, word in enumerate(vocab)}

# Convert sentences to numeric representation
X = np.array([[word_to_index[word] for word in sentence] for sentence in data])

# Define Autoencoder
input_layer = Input(shape=(X.shape[1],))
encoded = Dense(128, activation='relu')(input_layer)
decoded = Dense(X.shape[1], activation='softmax')(encoded)

autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

# Train the Autoencoder
autoencoder.fit(X, X, epochs=50)
```

Use the Hugging Face transformers library to fine-tune a pre-trained GPT-2
model on a custom text data and generate text

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer, Trainer, TrainingArguments
from datasets import load_dataset

# Load custom dataset (text file format)
dataset = load_dataset('text', data_files={'train': 'my_custom_data.txt'})

# Load pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['text'], return_tensors="pt", truncation=True, padding=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Fine-tune model with training arguments
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train']
)

trainer.train()

# Generate text
input_text = "Once upon a time"
inputs = tokenizer.encode(input_text, return_tensors="pt")
output = model.generate(inputs, max_length=100, num_return_sequences=1)
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

Implement a text generation model using a simple Recurrent Neural
Network (RNN) in Keras. Train the model on a custom data and generate a
word

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense, Embedding
import numpy as np

# Sample data (tokenized sentences)
data = ["The cat is on the mat", "The dog barks loudly", "Cats are cute"]
tokens = [sentence.split() for sentence in data]
vocab = list(set([word for sentence in tokens for word in sentence]))
word_to_index = {word: i for i, word in enumerate(vocab)}
index_to_word = {i: word for i, word in enumerate(vocab)}

# Prepare training data
X = np.array([[word_to_index[word] for word in sentence[:-1]] for sentence in tokens])
y = np.array([[word_to_index[word] for word in sentence[1:]] for sentence in tokens])

# Define RNN model
model = Sequential()
model.add(Embedding(len(vocab), 10, input_length=X.shape[1]))
model.add(SimpleRNN(50))
model.add(Dense(len(vocab), activation='softmax'))

# Compile and train the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=100)

# Text generation
def generate_text(model, seed_text, length=5):
    for _ in range(length):
        sequence = [word_to_index[word] for word in seed_text.split()]
        sequence = np.array(sequence).reshape(1, len(sequence))
        predicted_word_idx = np.argmax(model.predict(sequence))
        seed_text += " " + index_to_word[predicted_word_idx]
    return seed_text

generated_text = generate_text(model, "The cat", 5)
print(generated_text)
```
